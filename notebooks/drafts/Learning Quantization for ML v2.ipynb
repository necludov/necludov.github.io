# Classification Aware Lloyd-max Quantization

```python
import jax.numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import pandas as pd
import multiprocessing
import seaborn as sns

from joblib import Parallel, delayed
from IPython.display import Markdown, display
from numpy.random import seed as set_seed
from itertools import product

sns.set(style="darkgrid")
CPU_COUNT = multiprocessing.cpu_count()
parallel = Parallel(n_jobs=CPU_COUNT, verbose=3)
```

```python
def classify(S, w):
    return np.sign((S*w).sum(axis=1))

def make_normal_unit_vector(d):
    v = np.append(-1, np.ones(d-1))
    v = v/np.sqrt(v@v)
    return v[np.newaxis, :, np.newaxis]

def normalize(X):
    x = X - X.min(axis=0)
    return x/x.max(axis=0)

def reconstruction_loss(S, X) -> '(N, 2**R)':
    return ((S - X)**2).sum(axis=1)

def classification_loss(S, X, w) -> '(N, 2**R)':
    return classify(S, w) != classify(X, w)

def calc_total_loss(losses, γ_vec) -> '(N, 2**R)':
    return losses @ γ_vec

def codeword_losses(S, X, w) -> '(N, 2**R, 2)':
    return np.stack([reconstruction_loss(S, X),
                     classification_loss(S, X, w)], axis=-1)

def calc_optimal_Qinv_given_X(S, X, w, γ_vec) -> 'Qinv: (N,); losses: (2,)':
    losses = codeword_losses(S, X, w)
    loss_total = calc_total_loss(losses, γ_vec)
    Qinv: '(N,)' = loss_total.argmin(axis=1)
    return Qinv, losses[np.arange(S.shape[0]), Qinv].mean(axis=0)

def calc_optimal_X_given_Qinv(S, Qinv, w, γ_vec, R) -> '(N, d, 2**R)':
    X = np.array([S[Qinv==i].mean(axis=0)
                  for i in range(2**R)
                  if (Qinv==i).sum() > 0]).T
    ℓ_samples = calc_total_loss(codeword_losses(S, X, w), γ_vec)
    ℓ_codes = np.array([ℓ_samples[Qinv==i, i].mean()
                        for i in range(2**R)
                        if (Qinv==i).sum() > 0])
    Δ = -(X * w).sum(axis=1)*w*(1 + 1e-6)
    ℓΔ_samples = calc_total_loss(codeword_losses(S, X + Δ, w), γ_vec)
    ℓΔ_codes = np.array([ℓΔ_samples[Qinv==i, i].mean()
                         for i in range(2**R)
                         if (Qinv==i).sum() > 0])
    toggle_flag = np.stack((ℓ_codes, ℓΔ_codes)).argmin(axis=0)
    return X + toggle_flag*Δ

def lloyd_max_algorithm(S, X, w, γ_vec, R, max_iter=200):
    losses = max_iter*[0]
    for i in range(max_iter):
        Qinv, losses[i] = calc_optimal_Qinv_given_X(S, X, w, γ_vec)
        X = calc_optimal_X_given_Qinv(S, Qinv, w, γ_vec, R)
    return np.array(losses), X, Qinv

def plot(S, X, Qinv, w, title):
    wrong = (classify(S, w) != classify(X[:, :, Qinv], w).T).ravel()
    plt.scatter(*S.T[0], c=Qinv, alpha=0.8, cmap='Dark2')
    plt.scatter(*S[wrong].T[0], c='cyan', alpha=1)
    plt.scatter(*X[0], marker='X', c='black', s=100)
    plt.plot([0, 1], [0, 1], 'k--')
    plt.title(title)
    plt.axis('off')
```

All vectors and matrices are created with 3 dimensions: 
sample points, data dimension and number of reconstruction points

```python
seed: 'global numpy seed' = 1
set_seed(seed)

N: 'number of datapoints' = 2000
d: 'data dimension' = 2
R: 'rate (i.e. number of bits)' = 3
X: 'reconstruction points' = stats.uniform().rvs(size=(1, d, 2**R))
n_clusters: 'number of clusters used to generate S (must divide N)' = 10
S: 'sample' = (stats
               .multivariate_normal(mean=stats.uniform().rvs(n_clusters*d),
                                    cov=1e-2)
               .rvs(size=int(N/n_clusters))
               .reshape(-1, d)
               [:, :, np.newaxis])
# S = normalize(stats.norm().rvs(size=(N, d, 1)))
w: 'unit vector normal to decision boundary' = make_normal_unit_vector(d)
γ: 'classifiction (1) and reconstruction (0) trade-off'
losses: 'MSE and 0-1 losses for given X and Qinv'
Qinv: 'assigns a region to each datapoint'

assert S.shape == (N, d, 1)
assert X.shape == (1, d, 2**R)

def run_experiment(γ):
    γ_vec = np.array([1-γ, γ])
    losses, X_opt, Qinv_opt = lloyd_max_algorithm(S, X, w, γ_vec, R, max_iter=10)
    if d == 2:
        plt.figure(figsize=(20,5))
        plt.subplot(1,3,1)
        Qinv, _ = calc_optimal_Qinv_given_X(S, X, w, γ_vec)
        plot(S, X, Qinv, w, 'Initial quantizer')

        plt.subplot(1,3,2)
        plot(S, X_opt, Qinv_opt, w, 'Optimal quantizer (achieved via Lloyd-max)')

    (pd.DataFrame(losses, columns=['MSE loss', '0-1 loss'])
       .assign(**{'Total loss (MSE + 0-1)': losses.sum(axis=1)})
       .plot(ax=plt.subplot(1,3,3) if d == 2 else None,
             secondary_y='MSE loss',
             title='Reconstruction (MSE) and Classification (0-1) Losses',
             xlim=(0, losses.argmin(axis=0).max()),
             style='o--'))
    plt.show()

    return Markdown(f'''
$N={N}, d={d}, R={R}, \gamma={γ}$\n
Lowest reconstruction (MSE) loss: {losses.T[0].min():.5f} @ iter={losses.T[0].argmin()}\n
Lowest classification (0-1) loss: {losses.T[1].min():.5f} @ iter={losses.T[1].argmin()}\n
Lowest total (MSE + 0-1) loss: {losses.sum(axis=1).min():.5f} @ iter={losses.sum(axis=1).argmin()}\n
Final number of codewords: {X_opt.shape[2]}

---''')

for γ in [0, 1e-3, 1e-2, 1e-1, 0.95]:
    display(run_experiment(γ))
```

```python
try:
    metrics = pd.read_parquet('metrics.parq')
except:
    def single_run(R, d, γ):
        seed: 'global numpy seed' = 1
        set_seed(seed)
        N: 'number of datapoints' = 1000
        X: 'reconstruction points' = stats.uniform().rvs(size=(1, d, 2**R))
        n_clusters: 'number of clusters used to generate S (must divide N)' = 10
        S: 'sample' = (stats
                       .multivariate_normal(mean=stats.uniform().rvs(n_clusters*d),
                                            cov=1e-2)
                       .rvs(size=int(N/n_clusters))
                       .reshape(-1, d)
                       [:, :, np.newaxis])
        w: 'unit vector normal to decision boundary' = make_normal_unit_vector(d)
        γ_vec = np.array([1 - γ, γ])

        losses, _, _ = lloyd_max_algorithm(S, X, w, γ_vec, R, max_iter=50)
        i = (losses @ γ_vec).argmin()
        loss_total = (losses @ γ_vec)[i]
        loss_recon, loss_class = losses[i]

        return {'loss_total': loss_total.item(),
                'loss_recon': loss_recon.item(),
                'loss_class': loss_class.item(),
                'R': R,
                'd': d,
                'γ': γ}



    search_space = product([0, 1,  2,  3],     # R
                           [1, 5, 10],     # d
                           np.logspace(-3, 0, 20)) # γ

    metrics = (pd.DataFrame(parallel(delayed(single_run)(R, d, γ)
                                     for R, d, γ in search_space))
                 .melt(value_vars=['loss_total', 
                                   'loss_recon', 
                                   'loss_class'],
                       var_name='loss_type',
                       value_name='loss_value',
                       id_vars=['R', 'd', 'γ']))
metrics.head()
```

```python
replace = {
    'loss_type': {
        'loss_total': 'Total loss (MSE + 0-1)',
        'loss_recon': 'MSE loss',
        'loss_class': '0-1 loss',
    }
}

plt.figure(figsize=(22,15))

for i, loss_type in enumerate(metrics['loss_type'].unique(), start=1):
    
    (metrics
     .query("loss_type == @loss_type")
     .groupby(['d', 'R'])
     ['loss_value']
     .min()
     .unstack('R')
     .plot(ax=plt.subplot(3, 3, i),
           style='o--'))
    plt.title(replace['loss_type'][loss_type], fontsize=22)
    
    if i == 1:
        plt.ylabel('$\min_\gamma\ L_S^{(i)}(\gamma, R, d)$', fontsize=22)

    (metrics
     .query("loss_type == @loss_type")
     .groupby(['γ', 'd'])
     ['loss_value']
     .min()
     .unstack('d')
     .plot(ax=plt.subplot(3, 3, i+3),
           style='o--',
           logx=True))
    
    if i == 1:
        plt.ylabel('$\min_R\ L_S^{(i)}(\gamma, R, d)$', fontsize=22)

    (metrics
     .query("loss_type == @loss_type")
     .groupby(['γ', 'R'])
     ['loss_value']
     .min()
     .unstack('R')
     .plot(ax=plt.subplot(3, 3, i+6),
           style='o--',
           logx=True));
    
    if i == 1:
        plt.ylabel('$\min_d\ L_S^{(i)}(\gamma, R, d)$', fontsize=22)

plt.show()
Markdown('''
$L_S^{(i)}$ is the loss on sample $S$ at iteration $i$ which achieves the lowest Total loss.
''')
```
