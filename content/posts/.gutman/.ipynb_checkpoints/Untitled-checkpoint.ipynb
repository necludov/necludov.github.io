```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
import numpy as np

def normalize_densities(*args):
    return (np.array(P)/np.sum(P) for P in args)

def D(P, Q):
    '''KL Divergence'''
    p, q = normalize_densities(P, Q)
    kl = p*np.log2(p/q)
    return np.where(p == 0, 0, kl).sum()


def GJS(P, Q, α):
    '''Generalized Jensen-Shannon Divergence'''
    p, q = normalize_densities(P, Q)
    w = (α*p+q)/(1+α)
    return α*D(p,w) + D(q,w)

P = [1, 1]

assert D(P, P) == 0
assert all(GJS(P, P, α) for α in range(1, 10)) == 0
```

```python
P1 = [0.1, 0.7, 0.2]
P2 = [0.4, 0.7, 0.2]
P1, P2 = normalize_densities(P1, P2)
γ = 0.03
N = 300
n_space = np.arange(1, N+100)

Q_X1 = np.random.multinomial(N, P1)
Q_X2 = np.random.multinomial(N, P2)

Y = np.random.choice(len(P2), n_space.max(), p=P2)
Q_Y = np.array([[(Y[:n] == i).sum() for i in range(len(P2))] 
                for n in n_space])


score_1 = np.array([n*GJS(Q_X1, Q_Y[n-1], N/n) for n in n_space])
score_2 = np.array([n*GJS(Q_X2, Q_Y[n-1], N/n) for n in n_space])

plt.figure(figsize=(16,9))
plt.plot(n_space, score_1, n_space, score_2)
plt.hlines(γ*N, n_space.min(), n_space.max(), linestyles='--')
```
